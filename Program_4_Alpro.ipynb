```python
# Program untuk mengekstrak tanggal lahir dari NIP ASN
nip = input("Masukkan NIP ASN: ")

# Validasi panjang input minimal 8 digit awal
if len(nip) < 8:
    print("Error: NIP yang dimasukkan terlalu pendek (minimal 8 digit).")
else:
    # Proses ekstraksi komponen tanggal lahir
    tahun = nip[0:4]
    bulan_str = nip[4:6]
    tanggal = nip[6:8]

    # Percabangan untuk konversi nama bulan
    if bulan_str == "01":
        bulan = "Januari"
    elif bulan_str == "02":
        bulan = "Februari"
    elif bulan_str == "03":
        bulan = "Maret"
    elif bulan_str == "04":
        bulan = "April"
    elif bulan_str == "05":
        bulan = "Mei"
    elif bulan_str == "06":
        bulan = "Juni"
    elif bulan_str == "07":
        bulan = "Juli"
    elif bulan_str == "08":
        bulan = "Agustus"
    elif bulan_str == "09":
        bulan = "September"
    elif bulan_str == "10":
        bulan = "Oktober"
    elif bulan_str == "11":
        bulan = "November"
    elif bulan_str == "12":
        bulan = "Desember"
    else:
        bulan = "Tidak Valid"

    # Menampilkan hasil output program
    if bulan != "Tidak Valid":
        print(f"ASN tersebut lahir pada tanggal: {tanggal} {bulan} {tahun}")
    else:
        print("Error: Dua digit bulan pada NIP tidak valid.")
```
